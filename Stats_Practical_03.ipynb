```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
school_ages
```

```python
classA_ages
```

```python
classA_ages.mean()
```

```python
p_value=ttest_1samp(classA_ages,school_ages.mean())
```

```python
school_ages.mean()
```

```python
p_value_result=ttest_1samp(classA_ages,school_ages.mean())
p_value = p_value_result.pvalue
if p_value < 0.05:
  print("Reject HO")
else:
  print("Accept HO")
```
